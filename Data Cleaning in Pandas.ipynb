# Variables are stored in both rows and columns : Weather Dataset


```python
weather = pd.read_csv("weather.csv")
```

```python
weather
```

```python
#pivoting the rows || #DS360°withAkanksha | ©2020
#can use pd.melt, or weather.melt

weather_melt = weather.melt(
    
      id_vars= ['id','year','month','element'],
      var_name='day',
      value_name='temp'
)
```

```python
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(
    index=['id','year','month','day'],
    columns='element',
    values='temp'
)
```

```python
weather_tidy.reset_index()
```
